```python
import pandas as pd
import matplotlib.pyplot as plt
#defining the defination
df=pd.read_csv(r'C:\Users\Sandeep\Downloads\HDFC.csv',index_col='Date',parse_dates=True)
df.head()
```

```python
#shape of dataset
df.shape
```

```python
#drop the null values pertaining to column Prev Close

df=df.dropna()
df=df['Prev Close']['2019-01-01':'2019-12-31']
df=pd.DataFrame(df)
df.head()
```

```python
#ploting/visualizing the data

plt.figure(figsize=(16,10))
plt.plot(df)
plt.show()
```

```python
#making time series stationary

df1=df.diff(periods=1)
plt.figure(figsize=(15,10))
plt.plot(df1)
plt.show()
```
